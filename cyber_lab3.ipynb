```python
from scapy.all import rdpcap, TCP, Raw
from collections import defaultdict, Counter
import re

# Path to the pcap file
pcap_file_path = "/content/fake_email_traffic.pcap"

# Read packets from the pcap file
packets = rdpcap(pcap_file_path)

# Email regex pattern (SMTP, POP3, IMAP traffic)
email_pattern = re.compile(rb'[\w\.-]+@[\w\.-]+\.\w+')

# Dictionaries to store sender and receiver email addresses
senders = []
receivers = []
email_pairs = defaultdict(int)

# Process each packet
for packet in packets:
    if packet.haslayer(TCP) and packet.haslayer(Raw):  # Check if packet has TCP and Raw payload
        payload = packet[Raw].load

        # Extract email addresses from payload
        email_addresses = email_pattern.findall(payload)
        email_addresses = [email.decode() for email in set(email_addresses)]  # Decode bytes to string

        if len(email_addresses) >= 2:  # Expecting at least sender and receiver
            sender = email_addresses[0]
            receiver = email_addresses[1]

            senders.append(sender)
            receivers.append(receiver)
            email_pairs[(sender, receiver)] += 1

# Count most frequent senders and receivers
top_senders = Counter(senders).most_common(5)
top_receivers = Counter(receivers).most_common(5)
top_email_pairs = Counter(email_pairs).most_common(5)

# Print Report
print("\n **Top 5 Most Frequent Senders:**")
for sender, count in top_senders:
    print(f"{sender}: {count} emails sent")

print("\n **Top 5 Most Frequent Receivers:**")
for receiver, count in top_receivers:
    print(f"{receiver}: {count} emails received")

print("\n **Top 5 Sender-Receiver Pairs:**")
for (sender, receiver), count in top_email_pairs:
    print(f"{sender}  {receiver}: {count} emails exchanged")
```

```python
from scapy.all import rdpcap, TCP, Raw, IP  # Import required Scapy modules
from collections import defaultdict, Counter

# Path to the PCAP file
pcap_file_path = "/content/fake_email_traffic.pcap"

# Read packets from the PCAP file
packets = rdpcap(pcap_file_path)

# Storage for email traffic analysis
senders_to_recipients = defaultdict(set)  # {source IP: {destination IPs}}
recipients_from_senders = defaultdict(set)  # {destination IP: {source IPs}}
email_counts = Counter()  # Count of emails per source IP and per destination IP

# Process packets to extract SMTP traffic
for packet in packets:
    if packet.haslayer(IP) and packet.haslayer(TCP) and packet.haslayer(Raw):
        payload = packet[Raw].load
        if b"MAIL FROM" in payload or b"RCPT TO" in payload:  # Only process SMTP email packets
            src_ip = packet[IP].src
            dst_ip = packet[IP].dst

            senders_to_recipients[src_ip].add(dst_ip)  # Track unique recipients per sender
            recipients_from_senders[dst_ip].add(src_ip)  # Track unique senders per recipient
            email_counts[src_ip] += 1  # Count emails sent per source
            email_counts[dst_ip] += 1  # Count emails received per destination

# Identify suspicious activity
suspicious_senders = {ip: len(recipients) for ip, recipients in senders_to_recipients.items() if len(recipients) > 5}
suspicious_recipients = {ip: len(senders) for ip, senders in recipients_from_senders.items() if len(senders) > 5}
high_traffic_ips = {ip: count for ip, count in email_counts.items() if count > 10}

# Print suspicious findings
print("\n **Suspicious Senders (Sending to Many Recipients):**")
for ip, count in suspicious_senders.items():
    print(f" {ip} sent emails to {count} different recipients")

print("\n **Suspicious Recipients (Receiving from Many Senders):**")
for ip, count in suspicious_recipients.items():
    print(f" {ip} received emails from {count} different senders")

print("\n **High Traffic IPs (Unusual Email Volume):**")
for ip, count in high_traffic_ips.items():
    print(f" {ip} involved in {count} email transactions")
```

```python
from scapy.all import rdpcap, TCP, Raw, IP  # Import required Scapy modules
import re
from collections import defaultdict

# Path to the PCAP file
pcap_file_path = "/content/fake_email_traffic.pcap"

# Read packets from the PCAP file
packets = rdpcap(pcap_file_path)

# SMTP Command Patterns
smtp_regex = {
    "EHLO": re.compile(rb"EHLO"),
    "MAIL FROM": re.compile(rb"MAIL FROM:\s*<([^>]+)>"),
    "RCPT TO": re.compile(rb"RCPT TO:\s*<([^>]+)>"),
    "DATA": re.compile(rb"DATA"),
    "DOT": re.compile(rb"^\.\r\n"),  # End of DATA section
    "QUIT": re.compile(rb"QUIT")
}

# Storage for SMTP sessions
sessions = defaultdict(list)  # {source_ip: [commands]}
malformed_sessions = []  # List of problematic sessions

# Process packets to extract SMTP commands
for packet in packets:
    if packet.haslayer(IP) and packet.haslayer(TCP) and packet.haslayer(Raw):
        payload = packet[Raw].load
        src_ip = packet[IP].src
        dst_ip = packet[IP].dst

        for command, pattern in smtp_regex.items():
            if pattern.search(payload):  # Check if command is in payload
                sessions[src_ip].append(command)  # Store command in session

# Analyze SMTP sessions for irregularities
for src_ip, commands in sessions.items():
    if "MAIL FROM" in commands and "RCPT TO" not in commands:
        malformed_sessions.append((src_ip, "Missing RCPT TO after MAIL FROM"))

    if "DATA" in commands and "DOT" not in commands:
        malformed_sessions.append((src_ip, "DATA command without termination (missing '.')"))

    if "EHLO" in commands and "QUIT" not in commands:
        malformed_sessions.append((src_ip, "Session missing QUIT command"))

    if any(cmd not in ["EHLO", "MAIL FROM", "RCPT TO", "DATA", "DOT", "QUIT"] for cmd in commands):
        malformed_sessions.append((src_ip, "Unexpected SMTP command sequence"))

# Print the irregular SMTP sessions
print("\n **Malformed or Incomplete SMTP Sessions Detected:**")
if malformed_sessions:
    for src_ip, issue in malformed_sessions:
        print(f" {src_ip}: {issue}")
else:
    print(" No issues detected")
```
